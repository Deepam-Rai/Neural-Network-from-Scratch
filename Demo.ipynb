# Creating Binary Class Dataset using numpy

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random binary classification dataset
np.random.seed(123)
num_samples = 1000

#generate 2D features
x1 = np.random.normal(loc=-1.0, scale=1.0, size=num_samples)
x2 = np.random.normal(loc=1.0, scale=1.0, size=num_samples)
X = np.vstack([x1,x2]).T

#create labels
Y = ((x1 + x2 + np.random.normal(scale=0.5, size=num_samples)) > 0.0).astype(float)
# (x1 + x2>0) separates points on either side of line x+y=0 to one class each
# adding random noise overlaps the classes rendering linearly inseparable; just a little complexity


# Plot dataset
plt.scatter(x1[Y==0], x2[Y==0], alpha=0.5, label='Class 0')
plt.scatter(x1[Y==1], x2[Y==1], alpha=0.5, label='Class 1')
plt.legend()
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
print("X shape:",X.shape)
print("Y shape:", Y.shape)

# 1000 observations with 1000 labels
```

# Building Neural Network using Chochin
A simple feed forward neural network of:
- Input layer with 2 neurons
- 1 hidden layer with 10 neurons with ReLU activation
- Output layer with 1 neuron with Sigmoid Activation
  
 If the output>0.5 it is taken as class 1 else class 0.

```python
#hyper-parameters

input_dim = X.shape[1]
hidden_dim = 10
learning_rate = 0.01 #for SGD
num_epochs = 35
```

```python
class SampleNN(chochin.nn.NeuralNetwork):
  def __init__(self, input_dim, hidden_dim):
    super(SampleNN,self).__init__()
    
    self.hidden1 = chochin.nn.Linear(in_features=input_dim,out_features=hidden_dim)
    self.activation1 = chochin.activations.ReLU()
    
    self.final = chochin.nn.Linear(in_features=hidden_dim, out_features=1)
    self.final_activation = chochin.activations.Sigmoid()
  
  def forward(self,x):
    h = self.hidden1(x)
    h = self.activation1(h)
    h = self.final(h)
    h = self.final_activation(h)
    return h

model = SampleNN(input_dim,hidden_dim)
model
```

## Performance before training
Visualizing the classification by the network before training.

```python
#predict using untrained model
y_pred = model(X)
y_pred = (y_pred>0.5).astype(float)

# Plot dataset and color as per prediction
plt.scatter(x1[y_pred==0], x2[y_pred==0], alpha=0.5, label='Class 0')
plt.scatter(x1[y_pred==1], x2[y_pred==1], alpha=0.5, label='Class 1')
plt.legend()
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

# Training the model

## Loss and Optimizer
- Loss: BCELoss()
- Optimizer: Stochastic Gradient Descent.
  
Both classes used from chochin.

```python
lossFn = chochin.loss.BCELoss()

optimizer = chochin.optim.SGD(model,lr=0.05)

print("optimizer\n",optimizer)
```

```python
for epoch in range(num_epochs):
  running_loss = 0.0
  for i in range( len(X)):
    #get the prediction
    ypred = model.forward(X[i])
    #calculate the loss
    loss = lossFn(ypred, np.array([Y[i]]))
    #calculate the gradients
    model.backward(lossFn.backward())
    #update the parameters
    optimizer.step()

    running_loss += loss
  epoch_loss = running_loss/len(X)
  print(f"Epoch {epoch+1}/{num_epochs}, Loss: {epoch_loss.item():.4f}")
```

## Performance After training

```python
# predict using trained model
y_pred = model(X)
y_pred = (y_pred>0.5).astype(float)

# Plot dataset and color as per prediction
plt.scatter(x1[y_pred==0], x2[y_pred==0], alpha=0.5, label='Class 0')
plt.scatter(x1[y_pred==1], x2[y_pred==1], alpha=0.5, label='Class 1')
plt.legend()
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

# PyTorch network
We see the behavious of identical neural network built using pytorch for the same dataset.

```python
import torch
import torch.nn as nn

class BinaryRegressor(nn.Module):
    def __init__(self, input_dim, hidden_dim):
        super(BinaryRegressor, self).__init__()
        self.linear1 = nn.Linear(input_dim, hidden_dim)
        self.relu = nn.ReLU()
        self.linear2 = nn.Linear(hidden_dim, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        out = self.linear1(x)
        out = self.relu(out)
        out = self.linear2(out)
        out = self.sigmoid(out)
        return out
```

## Setting the data and instantiating model

```python
# Convert data to PyTorch tensors
inputs = torch.from_numpy(np.vstack((x1, x2)).T).float()
targets = torch.from_numpy(Y).float().reshape(-1, 1)


binary_regressor = BinaryRegressor(input_dim, hidden_dim)
```

## performance before training

```python
# predict using untrained pytorch model
with torch.no_grad():
    outputs = binary_regressor(inputs)
    predictions = (outputs > 0.5).float()

# Plot predictions and color as per predictions
plt.scatter(x1[predictions.flatten()==0], x2[predictions.flatten()==0], alpha=0.5, label='Class 0')
plt.scatter(x1[predictions.flatten()==1], x2[predictions.flatten()==1], alpha=0.5, label='Class 1')
plt.legend()
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

## Training pytorch network

```python
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(binary_regressor.parameters(), lr=learning_rate)

# Training loop
for epoch in range(num_epochs):
    running_loss = 0.0
    for i in range(num_samples):
      # clear old gradients
      optimizer.zero_grad()
      # predict
      outputs = binary_regressor(inputs[i])
      #calculate loss
      loss = criterion(outputs, targets[i])
      # calculate gradients
      loss.backward()
      #update the parameters
      optimizer.step()
      running_loss += loss.item()
    epoch_loss = running_loss / num_samples
    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {epoch_loss:.4f}")
```

## Performance after training

```python
# Evaluate trained model on the dataset
binary_regressor.eval()
with torch.no_grad():
    outputs = binary_regressor(inputs)
    predictions = (outputs > 0.5).float()

# Plot predictions and color as per predictions
plt.scatter(x1[predictions.flatten()==0], x2[predictions.flatten()==0], alpha=0.5, label='Class 0')
plt.scatter(x1[predictions.flatten()==1], x2[predictions.flatten()==1], alpha=0.5, label='Class 1')
plt.legend()
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```
